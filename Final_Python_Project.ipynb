```python
import pandas as pd
import numpy as np
```

```python
#Load datasets from csv file

df_sales_data = pd.read_csv("zepto_sales.csv")
df_product_data = pd.read_csv("zepto_sales.csv")
```

```python
#display datasets
print(df_product_data)
print(df_sales_data)
```

```python
# display basic information of sales data
print(df_sales_data.info())
print(df_sales_data.describe())
print(df_sales_data.shape)
```

```python
# display bsic information of products data
print(df_product_data.info())
print(df_product_data.describe())
print(df_product_data.shape)
```

```python
# display top records of sales and products data 
print(df_product_data.head())
print(df_sales_data.head())
```

```python
# display last records of sales and proucts data
print(df_sales_data.tail())
print(df_product_data.tail())
```

```python
# check null values in sales data

print(df_sales_data.isnull().sum())
```

```python
# remove null values from city and delivery_status column

df_sales_data.dropna(subset=["city","delivery_status"],inplace=True)
```

```python
# After this again check null values
print(df_sales_data[["city","delivery_status"]].isnull().sum())
```

```python
# check null values in delivery_time_mins column and fill these values with mean of delivery_time_mins

print(df_sales_data.isnull().sum())
mean_delivery_time = df_sales_data["delivery_time_mins"].mean()
print(mean_delivery_time)
df_sales_data.fillna({"delivery_time_mins":mean_delivery_time}, inplace=True)
```

```python
# again checking null values 
print(df_sales_data["delivery_time_mins"].isnull().sum())
```

```python
# checking duplicate records

print(f"number of duplicate rows is:{df_sales_data.duplicated().sum()}")
```

```python
# remove duplicate records and print number of records 

df_sales_data.drop_duplicates(inplace=True)
print(f"number of records after removing duplicate records is : {df_sales_data.shape[0]}")
```

```python
# convert order date to date time objects

df_sales_data["order_date"]=pd.to_datetime(df_sales_data["order_date"])
```

```python
# print columns name of sales data
df_sales_data.columns
```

```python
# find min max Average of total amount

min_amount = df_sales_data["total_amount"].min()
max_amount = df_sales_data["total_amount"].max()
Average_amount = df_sales_data["total_amount"].mean()

print("Min of total amount is: ",min_amount)
print("Max of total amount is: ", max_amount)
print("Average of total amount is : ", Average_amount)
```

```python
# Top 10 products by total sales amount

Top_products = df_sales_data.groupby("product_id")["total_amount"].sum().nlargest(10)
print("--Top 10 products by sales amount--")
print(Top_products)
```

```python
# Merge with product details to get product names

top_products_deatils= Top_products.reset_index().merge(df_product_data, on= "product_id")
print("---top products details---")
print(top_products_deatils)
```

```python
# total sales by city only top 5

total_sales_by_city= df_sales_data.groupby("city")["total_amount"].sum().sort_values(ascending=False).head(5)
print("---Total sales by city---")
print(total_sales_by_city)
```

```python
# Average delivery time by city

Avg_delivery_time_by_city= df_sales_data.groupby("city")["delivery_time_mins"].mean().sort_values(ascending=False).head(5)
print("---Average delivery time by city top 5---")
print(Avg_delivery_time_by_city)
```

```python
# Sales trend over time (e.g., monthly sales)

df_sales_data["month"] = df_sales_data["order_date"].dt.to_period("M")
print(df_sales_data["month"])
monthly_sales = df_sales_data.groupby("month")["total_amount"].sum()
print("monthly_sales data", monthly_sales)
```

```python
# Data visualization with matplolib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
sns.set_theme("paper")

plt.rcParams["figure.figsize"] = (10,8)
```

```python
# plot 1: top sales by city

plt.figure(figsize=(10,6))
sns.barplot(x=total_sales_by_city.index, y=total_sales_by_city.values, palette="magma")
plt.title("total sales by city")
plt.xlabel("city")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# plot 2: monthly sales 

plt.figure(figsize=(10, 7))
monthly_sales.plot(kind="line", marker="o",color="#3498DB")
plt.title("monthly_sales data")
plt.xlabel("Month")
plt.ylabel("total sales amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# plot 3: Distribution of Delivery Times


plt.figure(figsize=(10, 6))
sns.histplot(df_sales_data["delivery_time_mins"], bins=20, kde=True, 
color="#2ECC71")
plt.title("Distribution of Delivery Times")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
